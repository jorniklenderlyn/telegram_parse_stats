```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import numpy as np
import evaluate  # pip install evaluate
from sklearn.metrics import accuracy_score, classification_report
```

```python
# @title
labels = [
    "Frontend",
    "Backend",
    "Fullstack",
    "DevOps",
    "Data Scientist",
    "ML Engineer",
    "QA",
    "Product Manager"
    ]
num_labels = len(labels)
id2label = {i: label for i, label in enumerate(labels)}
label2id = {label: i for i, label in enumerate(labels)}

model_name = "DeepPavlov/rubert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels, id2label=id2label, label2id=label2id)

train_texts = [
    "React Vue TypeScript UI UX Figma",
    "Python Django FastAPI PostgreSQL Docker backend",
    "Fullstack React Node MongoDB",
    "Kubernetes CI/CD Jenkins AWS DevOps",
    "Pandas Tableau SQL Data Scientist –∞–Ω–∞–ª–∏–∑",
    "TensorFlow PyTorch MLflow ML Engineer",
    "Selenium pytest QA —Ç–µ—Å—Ç—ã",
    "Agile Scrum roadmap Product Manager"
]

train_labels = [0,1,2,3,4,5,6,7]

test_texts = ["JavaScript HTML CSS frontend —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫", "Golang –º–∏–∫—Ä–æ—Å–µ—Ä–≤–∏—Å—ã backend", "Data analysis Python ML"]
test_labels = [0,1,4]

train_data = {"text": train_texts, "labels": train_labels}
test_data = {"text": test_texts, "labels": test_labels}

train_dataset = Dataset.from_dict(train_data)
test_dataset = Dataset.from_dict(test_data)

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=256)

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(
    output_dir="./it-professions-model",
    num_train_epochs=3 * 15,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=20,
    weight_decay=0.01,
    logging_dir="./logs",
    eval_strategy="no",
    save_strategy="epoch",
    load_best_model_at_end=False,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    compute_metrics=compute_metrics
)

trainer.train()
eval_results = trainer.evaluate()
print("Eval results:", eval_results)

trainer.save_model("./it-professions-model")
tokenizer.save_pretrained("./it-professions-model")
```

```python
from ultralytics import YOLO
import cv2


model = YOLO('models/yolov8n.pt', task='classify')


def classify_image(image_path: str) -> tuple[str, float]:
    results = model(image_path)  # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ 
    
    top_class_id = results[0].probs.top1  # ID —Ç–æ–ø-–∫–ª–∞—Å—Å–∞
    top_conf = results[0].probs.top1conf.item()  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å
    class_name = model.names[int(top_class_id)]  # –ù–∞–∑–≤–∞–Ω–∏–µ –∫–ª–∞—Å—Å–∞
    
    return class_name, top_conf
```

```python
from transformers import pipeline

profession_classifier = pipeline("text-classification", model="./it-professions-model", tokenizer="./it-professions-model", return_all_scores=True)
```

```python
# stop words filtration
nlp = spacy.load("ru_core_news_sm")

def remove_stopwords(text: str) -> list[str]:
    doc = nlp(text)
    return ' '.join([token.text for token in doc if not token.is_stop and token.is_alpha])

def lemmatize(text: str) -> str:
    doc = nlp(text)
    return ' '.join([token.lemma_ for token in doc])


def classify_profession(text):
  result = profession_classifier(text)
  top_preds = sorted(result[0], key=lambda x: x['score'], reverse=True)[:3]
  for pred in top_preds:
      print(f"{pred['label']}: {pred['score']:.2%}")


morph = pymorphy3.MorphAnalyzer()

def normalize_to_full(name_str):
    if not name_str:
        return name_str

    if name_str.isascii():  # –¢–æ–ª—å–∫–æ –ª–∞—Ç–∏–Ω—Å–∫–∏–µ —Å–∏–º–≤–æ–ª—ã
        name_str = translit(name_str, 'ru') 

    words = name_str.strip().split()
    full_parts = []
    for word in words:
        parsed = morph.parse(word)[0]
        normal = parsed.normal_form.title()
        full_parts.append(normal)
    return ' '.join(full_parts)


async def get_user_full_info(client, user):
    try:
        full = await client(GetFullUserRequest(user))
        return {
            'about': full.full_user.about or '',
            'phone': getattr(full.full_user.common_chats_participants or [], 'phone_calls', None),  # –ù–µ –≤—Å–µ–≥–¥–∞
            'full': full.full_user
        }
    except (ChatAdminRequiredError, UserPrivacyRestrictedError):
        return {'about': '', 'phone': None}
    except Exception:
        return {'about': '', 'phone': None}

def process_text_for_ngrams(text):
    if not text:
        return []
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.text.strip()]
    unigrams = tokens
    # –î–ª—è –±–∏–≥—Ä–∞–º–º –∏ —Ç—Ä–∏–≥—Ä–∞–º–º (ngrams)
    bigrams = [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
    trigrams = [' '.join(tokens[i:i+3]) for i in range(len(tokens)-2)]
    return unigrams + bigrams + trigrams
```

```python
MAX_RETRIES = 3
SAVE_INTERVAL = 50

def setup_logging(group_id: int):
    avatar_dir = f"avatar/{group_id}"
    os.makedirs(avatar_dir, exist_ok=True)
    
    log_file = f"{avatar_dir}/scraper.log"
    
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s [%(levelname)s] %(message)s',
        handlers=[
            logging.FileHandler(log_file, encoding='utf-8'),
        ]
    )
    logger = logging.getLogger(__name__)
    return logger


async def download_user_photo(client, group_id: int, user_id: int, logger) -> bool:
    avatar_dir = f"avatar/{group_id}"
    filename = f"{avatar_dir}/{user_id}.jpg"
    
    if os.path.exists(filename):
        logger.info(f"‚è≠Ô∏è –ê–≤–∞—Ç–∞—Ä {user_id} —É–∂–µ —Å–∫–∞—á–∞–Ω")
        return True
    
    for attempt in range(MAX_RETRIES):
        try:
            photos = await client(GetUserPhotosRequest(user_id=user_id, offset=0, max_id=0, limit=1))
            if photos.photos:
                photo = photos.photos[0]
                await client.download_media(photo, filename)
                logger.info(f"üì∏ –ê–≤–∞—Ç–∞—Ä {user_id} ‚Üí {filename}")
                return True
            logger.info(f"‚ÑπÔ∏è –ê–≤–∞—Ç–∞—Ä {user_id} –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç")
            return True
        except FloodWaitError as e:
            logger.warning(f"‚è≥ Flood —Ñ–æ—Ç–æ {user_id}: {e.seconds}s (–ø–æ–ø—ã—Ç–∫–∞ {attempt+1})")
            await asyncio.sleep(e.seconds + 1)
        except Exception as e:
            logger.warning(f"üîÑ –§–æ—Ç–æ {user_id}: {e} (–ø–æ–ø—ã—Ç–∫–∞ {attempt+1}/{MAX_RETRIES})")
            if attempt < MAX_RETRIES - 1:
                await asyncio.sleep(2 ** attempt)
    
    logger.error(f"üí• –§–æ—Ç–æ {user_id}: –ª–∏–º–∏—Ç –ø–æ–ø—ã—Ç–æ–∫")
    return False


async def load_existing_about(avatar_dir: str, logger) -> dict:
    """–ó–∞–≥—Ä—É–∂–∞–µ—Ç —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–π —Ñ–∞–π–ª about –µ—Å–ª–∏ –µ—Å—Ç—å"""
    json_path = f"{avatar_dir}/users_description.json"
    if os.path.exists(json_path):
        try:
            with open(json_path, 'r', encoding='utf-8') as f:
                existing_about = json.load(f)
            logger.info(f"üìÇ –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(existing_about)} –∑–∞–ø–∏—Å–µ–π about –∏–∑ {json_path}")
            return {int(k): v for k, v in existing_about.items()}  # –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –∫–ª—é—á–∏ –≤ int
        except Exception as e:
            logger.warning(f"‚ö†Ô∏è –û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ {json_path}: {e}")
    
    logger.info("üì≠ –§–∞–π–ª about –Ω–µ –Ω–∞–π–¥–µ–Ω, –Ω–∞—á–∏–Ω–∞–µ–º —Å –ø—É—Å—Ç–æ–≥–æ")
    return {}


async def save_about_batch(all_about: dict, avatar_dir: str, count: int, logger):
    """–°–æ—Ö—Ä–∞–Ω—è–µ—Ç about –∫–∞–∂–¥—ã–µ 50 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π"""
    json_path = f"{avatar_dir}/users_description.json"
    try:
        with open(json_path, 'w', encoding='utf-8') as f:
            json.dump(all_about, f, ensure_ascii=False, indent=2)
        logger.info(f"üíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–æ {len(all_about)} –∑–∞–ø–∏—Å–µ–π ({count} –≤—Å–µ–≥–æ) ‚Üí {json_path}")
    except Exception as e:
        logger.error(f"üí• –û—à–∏–±–∫–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è {json_path}: {e}")


async def get_user_about(client, user_id: int, existing_about: dict, logger) -> dict:
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç about —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –µ–≥–æ –Ω–µ—Ç –≤ existing_about"""
    user_id_str = str(user_id)
    
    # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –µ—Å–ª–∏ about —É–∂–µ –µ—Å—Ç—å
    if user_id_str in existing_about:
        logger.info(f"‚è≠Ô∏è About {user_id} —É–∂–µ –µ—Å—Ç—å, –ø—Ä–æ–ø—É—Å–∫–∞–µ–º")
        return {}
    
    for attempt in range(MAX_RETRIES):
        try:
            full_user = await client(GetFullUserRequest(user_id))
            about = full_user.full_user.about or ""
            logger.debug(f"üìù About {user_id}: {about[:50]}...")
            return {
                user_id_str: {
                    "username": full_user.users[0].username or "",
                    "first_name": full_user.users[0].first_name or "",
                    "last_name": full_user.users[0].last_name or "",
                    "about": about
                }
            }
        except FloodWaitError as e:
            logger.warning(f"‚è≥ Flood about {user_id}: {e.seconds}s (–ø–æ–ø—ã—Ç–∫–∞ {attempt+1})")
            await asyncio.sleep(e.seconds + 1)
        except UserPrivacyRestrictedError:
            logger.warning(f"üîí About {user_id}: –ø—Ä–∏–≤–∞—Ç–Ω–æ—Å—Ç—å")
            return {}
        except ChatAdminRequiredError:
            logger.warning(f"üö´ About {user_id}: –Ω—É–∂–Ω—ã –ø—Ä–∞–≤–∞ –∞–¥–º–∏–Ω–∞")
            return {}
        except Exception as e:
            logger.warning(f"üîÑ About {user_id}: {e} (–ø–æ–ø—ã—Ç–∫–∞ {attempt+1}/{MAX_RETRIES})")
            if attempt < MAX_RETRIES - 1:
                await asyncio.sleep(2 ** attempt)
    
    logger.error(f"üí• About {user_id}: –ª–∏–º–∏—Ç –ø–æ–ø—ã—Ç–æ–∫")
    return {}


# –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≥—Ä—É–ø–ø
for group_name, group_id in GROUPS.items():
    logger = setup_logging(group_id)
    
    avatar_dir = f"avatar/"
    all_about = await load_existing_about(avatar_dir, logger)  # –ó–∞–≥—Ä—É–∂–∞–µ–º —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ
    count, photo_success, about_success, new_about_count = 0, 0, 0, 0
    
    logger.info(f"üöÄ –ì—Ä—É–ø–ø–∞ {group_id} ‚Üí {avatar_dir}/ (–Ω–∞–π–¥–µ–Ω–æ {len(all_about)} –∑–∞–ø–∏—Å–µ–π)")
    
    async for user in client.iter_participants(group_id):
        user_id = user.id
        
        # –°–∫–∞—á–∏–≤–∞–µ–º —Ñ–æ—Ç–æ (–≤—Å–µ–≥–¥–∞, –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç about)
        photo_task = await download_user_photo(client, group_id, user_id, logger)
        
        # –ü–æ–ª—É—á–∞–µ–º about —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –µ–≥–æ –Ω–µ—Ç
        user_about = await get_user_about(client, user_id, all_about, logger)
        
        if user_about:
            all_about.update(user_about)
            about_success += 1
            new_about_count += 1
        
        if photo_task:
            photo_success += 1
        
        count += 1
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∫–∞–∂–¥—ã–µ 50 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
        if count % SAVE_INTERVAL == 0:
            await save_about_batch(all_about, avatar_dir, count, logger)
            logger.info(f"üìä {count}: —Ñ–æ—Ç–æ={photo_success}, about={about_success} (+{new_about_count} –Ω–æ–≤—ã—Ö)")
    
    # –§–∏–Ω–∞–ª—å–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ
    await save_about_batch(all_about, avatar_dir, count, logger)
    
    logger.info(f"üéâ –ì–û–¢–û–í–û! {len(all_about)} –≤—Å–µ–≥–æ about ({new_about_count} –Ω–æ–≤—ã—Ö) ‚Üí {avatar_dir}/users_description.json")
    logger.info(f"üìà –ò–¢–û–ì–û: —Ñ–æ—Ç–æ {photo_success}/{count}, about {about_success}/{count}")
```

```python
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from typing import List, Dict
```

```python
def preprocess_text(text: str) -> str:
    spacy_stopwords = nlp.Defaults.stop_words
    doc = nlp(text)
    tokens = [
        token.lemma_.lower() 
        for token in doc 
        if (token not in spacy_stopwords and 
            not token.is_stop and 
            not token.is_punct and 
            token.is_alpha)
    ]
    return ' '.join(tokens)


def train_vectorizer(train_texts: List[str], max_features: int = 50, ngram_range: tuple = (1, 3)) -> TfidfVectorizer:
    train_processed = [preprocess_text(text) for text in train_texts]
    spacy_stopwords = list(nlp.Defaults.stop_words)
    vectorizer = TfidfVectorizer(
        ngram_range=ngram_range, 
        max_features=max_features, 
        stop_words=spacy_stopwords
    )
    vectorizer.fit(train_processed)
    return vectorizer

def extract_keywords(vectorizer: TfidfVectorizer, texts: List[str], top_n: int = 10) -> List[Dict]:
    processed_texts = [preprocess_text(text) for text in texts]
    tfidf_matrix = vectorizer.transform(processed_texts)
    feature_names = vectorizer.get_feature_names_out()
    results = []
    for i, post_scores in enumerate(tfidf_matrix.toarray()):
        top_indices = post_scores.argsort()[-top_n:][::-1]
        keywords = [feature_names[idx] for idx in top_indices if post_scores[idx] > 0 and feature_names[idx] in processed_texts[i]]
        results.append({"post_index": i, "top_keywords": keywords[:top_n]})
    return results
```

```python
train_texts = [post[-1].message for post in last_100_posts.values()]

texts = [post[-1].message for post in popular_posts]
```

```python
# –û–±—É—á–∞–µ–º
vectorizer = train_vectorizer(train_texts, max_features=None, ngram_range=(1, 3))
```

```python
# –ò–∑–≤–ª–µ–∫–∞–µ–º –∫–ª—é—á–µ–≤—ã–µ —Å–ª–æ–≤–∞
keywords_per_post = extract_keywords(vectorizer, texts[:], top_n=10)
for item in keywords_per_post:
    print(f"–ü–æ—Å—Ç {item['post_index']}: {item['top_keywords']}")
```

```python
import json
from jinja2 import Environment, BaseLoader

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
with open('first_part.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

groups_stats = data['groups_stats']

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —à–∞–±–ª–æ–Ω–∞
processed_groups = []
for group_name, stats in groups_stats.items():
    # –¢–æ–ø-5 –∏–º—ë–Ω
    top_names = sorted(stats['names'].items(), key=lambda x: x[1], reverse=True)[:5]
    top_names_labels = [name.title() for name, _ in top_names]
    top_names_values = [count for _, count in top_names]

    # –¢–æ–ø-5 –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
    top_active = sorted(
        stats['users_activity'].items(),
        key=lambda x: x[1]['messages'],
        reverse=True
    )[:5]
    top_active_users = []
    top_active_counts = []
    for user_id, info in top_active:
        display_name = info['username'] or f"{info['first_name']} {info['last_name'][0]}."
        top_active_users.append(display_name)
        top_active_counts.append(info['messages'])

    # –¢–æ–ø-5 n-–≥—Ä–∞–º–º
    top_ngrams = sorted(
        stats['messages_ngrams'].items(),
        key=lambda x: x[1],
        reverse=True
    )[:5]
    ngram_labels = [ng for ng, _ in top_ngrams]
    ngram_values = [cnt for _, cnt in top_ngrams]

    processed_groups.append({
        'name': group_name,
        'users_quantity': stats['users_quantity'],
        'users_with_open_numbers': stats['users_with_open_numbers'],
        'users_with_premium': stats['users_with_premium'],
        'percent_open': round(stats['users_with_open_numbers'] / stats['users_quantity'] * 100, 1),
        'percent_premium': round(stats['users_with_premium'] / stats['users_quantity'] * 100, 1),
        'top_names_labels': top_names_labels,
        'top_names_values': top_names_values,
        'top_active_users': top_active_users,
        'top_active_counts': top_active_counts,
        'ngram_labels': ngram_labels,
        'ngram_values': ngram_values
    })

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ Jinja2
env = Environment(loader=BaseLoader())

def format_number(value):
    return f"{value:,}".replace(",", " ")

env.filters['format_number'] = format_number

# –®–∞–±–ª–æ–Ω –¥–∞—à–±–æ—Ä–¥–∞
template_str = '''
<!DOCTYPE html>
<html lang="ru">
<head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1" />
    <title>–ê–Ω–∞–ª–∏—Ç–∏–∫–∞ Telegram-–≥—Ä—É–ø–ø</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.2/dist/css/bootstrap.min.css" rel="stylesheet" />
    <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap-icons@1.11.3/font/bootstrap-icons.css" />
    <style>
        .group-card { border-left: 4px solid #0d6efd; margin-bottom: 24px; }
        .stat-badge {
            padding: 0.4rem 0.8rem;
            font-weight: 600;
            border-radius: 8px;
            margin: 2px;
            display: inline-block;
        }
        .bg-premium { background-color: #e8f5e9; color: #2e7d32; }
        .bg-open { background-color: #fff8e1; color: #ef6c00; }
        .bg-total { background-color: #e3f2fd; color: #1565c0; }
        .keyword-tag {
            background: #f1f3f5;
            padding: 4px 10px;
            border-radius: 12px;
            margin: 2px;
            font-size: 0.875rem;
        }
        .chart-placeholder {
            background: #f8f9fa;
            height: 120px;
            display: flex;
            align-items: center;
            justify-content: center;
            border: 1px dashed #ccc;
            border-radius: 6px;
            font-size: 0.9rem;
            color: #666;
        }
    </style>
</head>
<body>
    <nav class="navbar navbar-dark bg-primary mb-4">
        <div class="container">
            <span class="navbar-brand">üìä –ê–Ω–∞–ª–∏—Ç–∏–∫–∞ Telegram-–≥—Ä—É–ø–ø</span>
        </div>
    </nav>

    <div class="container">
        {% for group in groups %}
        <div class="card group-card shadow-sm">
            <div class="card-header bg-light d-flex justify-content-between align-items-center">
                <h5 class="mb-0">{{ group.name }}</h5>
                <span class="badge bg-primary">{{ group.users_quantity | format_number }} —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤</span>
            </div>
            <div class="card-body">
                <div class="row mb-3">
                    <div class="col-md-4">
                        <div class="stat-badge bg-total">üë• –í—Å–µ–≥–æ: {{ group.users_quantity | format_number }}</div>
                    </div>
                    <div class="col-md-4">
                        <div class="stat-badge bg-premium">‚≠ê Premium: {{ group.users_with_premium | format_number }} ({{ group.percent_premium }}%)</div>
                    </div>
                    <div class="col-md-4">
                        <div class="stat-badge bg-open">üìû –û—Ç–∫—Ä—ã—Ç—ã–µ –Ω–æ–º–µ—Ä–∞: {{ group.users_with_open_numbers | format_number }} ({{ group.percent_open }}%)</div>
                    </div>
                </div>

                <div class="row">
                    <div class="col-md-6 mb-3">
                        <h6>–¢–æ–ø –∏–º—ë–Ω</h6>
                        {% for name in group.top_names_labels %}
                        <span class="keyword-tag">{{ name }}</span>
                        {% endfor %}
                    </div>
                    <div class="col-md-6 mb-3">
                        <h6>–ê–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏</h6>
                        {% for user in group.top_active_users %}
                        <div><small>{{ user }}</small></div>
                        {% endfor %}
                    </div>
                </div>

                <div class="mt-3">
                    <h6>–ü–æ–ø—É–ª—è—Ä–Ω—ã–µ —Ç–µ–º—ã (n-–≥—Ä–∞–º–º—ã)</h6>
                    {% for ng in group.ngram_labels %}
                    <span class="keyword-tag">{{ ng }}</span>
                    {% endfor %}
                </div>
            </div>
        </div>
        {% endfor %}

        <!-- –ì—Ä–∞—Ñ —Å–≤—è–∑–µ–π -->
        <div class="card mt-5 shadow-sm">
            <div class="card-header bg-dark text-white">
                <h5 class="mb-0"><i class="bi bi-diagram-3"></i> –ì—Ä–∞—Ñ —Å–≤—è–∑–µ–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏</h5>
            </div>
            <div class="card-body text-center">
                <img src="groups_graph.png" alt="–ì—Ä–∞—Ñ —Å–≤—è–∑–µ–π –≥—Ä—É–ø–ø" class="img-fluid border rounded" style="max-height: 800px;">
                <p class="text-muted mt-2">–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏</p>
            </div>
        </div>

        <footer class="text-center text-muted py-4 mt-5 border-top">
            –î–∞–Ω–Ω—ã–µ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –∞–Ω–∞–ª–∏–∑–∞ Telegram-–≥—Ä—É–ø–ø ‚Ä¢ {{ groups | length }} –≥—Ä—É–ø–ø
        </footer>
    </div>

    <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.3.2/dist/js/bootstrap.bundle.min.js"></script>
</body>
</html>
'''

# –†–µ–Ω–¥–µ—Ä–∏–Ω–≥
template = env.from_string(template_str)
html_output = template.render(groups=processed_groups)

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ
with open('groups_dashboard.html', 'w', encoding='utf-8') as f:
    f.write(html_output)

print("‚úÖ –î–∞—à–±–æ—Ä–¥ –≥—Ä—É–ø–ø —Å–æ–∑–¥–∞–Ω: groups_dashboard.html")
print("üí° –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ —Ñ–∞–π–ª 'groups_graph.png' –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —Ç–æ–π –∂–µ –ø–∞–ø–∫–µ!")
```
